## Content-based movie recommender system


This type of filter does not involve other users if not ourselves. Based on what we like, the algorithm will simply pick items with similar content to recommend us.

```python
#importing Libraries
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
```

Since , we have only 3 movies with missing overview we can simply drop them.

```python
movies.dropna(inplace = True)
```

### Preprocessing genres column.

```python
movies.head(10)
```

```python
(movies.iloc[0].genres)
```

Since genres are in string format we have to convert them , for further processing

```python
import ast 
def process(genre):
    final = []
    for i in ast.literal_eval(genre):
        final.append(i['name'])
    
    return final
```

```python
movies['genres'] = movies['genres'].apply(process)
```

```python
movies
```

### Preprocessing keywords column.

```python
movies['keywords'] = movies['keywords'].apply(process)
```

```python
movies.head(10)
```

### Preprocessing cast column.

```python
movies['cast'][0]
```

```python
# we will be extracting top 3 actors who worked in that movie
def actors(cast):
    final = []
    count = 0
    for i in ast.literal_eval(cast):
        if(count<3):
            final.append(i['name'])
            count+=1
        else:
            break
    
    return final
```

```python
movies['cast'] = movies['cast'].apply(actors)
```

```python
movies.head(10)
```

### Preprocessing crew column.

From all the crew members , we are only interested in knowing the director of the movie , because people may like movies of same the same director.

```python
movies['crew'][0]
```

```python
def director(crew):
    final = []
    for i in ast.literal_eval(crew):
        if i['job'] == 'Director':
            final.append(i['name'])
    
    return final
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies.head(10)
```

### Preprocessing overview column.

```python
#overview is in string format so we will be converting it into list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(10)
```

Before we move any further , one thing we have to do is to remove spaces between words. For example Christopher Nolan , should be replaced by ChristopherNolan , so our algorithm is not confused by similar prefix or suffix.

```python
# processing genres
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" " ,"") for i in x])
```

```python
# processing keywords
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" " ,"") for i in x])
```

```python
# processing cast
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" " ,"") for i in x])
```

```python
# processing crew
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" " ,"") for i in x])
```

```python
movies.head(10)
```

```python
# Now we have to concatenate all the columns to make a new columns which will have all the tags of a movie
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
len(movies)
```

Now we dont require the rest of the columns so we can create a new dataFrame

```python
df = movies[['movie_id' , 'title' , 'tags']]
```

```python
df
```

```python
#Converting Tags column into a string
df['tags'] = df['tags'].apply(lambda x:" ".join(x))
```

```python
# printing first tag
df['tags'][0]
```

```python
#converting string of tags columns into lower case
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
# printing first tag
df['tags'][0]
```

```python
len(df)
```

## Cleaning Texts

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []
for i in df['tags']:
  #removing all Punctuations.
  tag = re.sub('[^a-zA-Z]', ' ', i)
 
  #Splitting tags into different Words.
  tag = tag.split()
  
  #Stemming.
  ps = PorterStemmer()
  all_stopwords = stopwords.words('english')
  all_stopwords.remove('not')
  tag = [ps.stem(word) for word in tag if not word in set(all_stopwords)]
  #Now since we have applied stemming to all the words we again join them back
  tag   = ' '.join(tag)
  corpus.append(tag)
```

```python
print(corpus)
```

## Bag of Word Model

```python
from sklearn.feature_extraction.text import CountVectorizer
#limiting the number of different words to 5000
cv = CountVectorizer(max_features = 5000)
vectors = cv.fit_transform(corpus).toarray()
```

```python
# sparse Matrix
vectors
```

```python
#list of 5000 most common words in our corpus
cv.get_feature_names()
```

## Cosine Similarity

Cosine similarity is a metric used to measure how similar the documents are irrespective of their size. Mathematically, it measures the cosine of the angle between two vectors projected in a multi-dimensional space. The cosine similarity is advantageous because even if the two similar documents are far apart by the Euclidean distance (due to the size of the document), chances are they may still be oriented closer together. The smaller the angle, higher the cosine similarity.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_matrix = cosine_similarity(vectors)
```

```python
similarity_matrix
```

```python
# now we will recommend 5 movies which has similarity score near to 1
def recommendation_system(movie):
    try:
        movie_index = df[df['title'] == movie].index[0]
        cosine_angles = similarity_matrix[movie_index]
        recommended_movies = sorted(list(enumerate(cosine_angles)) , reverse = True , key = lambda x:x[1])[1:6]
        print("Top 5 recommended movies based on your choice:")
        print()
        k = 1
        for i in recommended_movies:
            print(k , "->", df.iloc[i[0]].title)
            k+=1
    except:
        print("This movie is not in our DataBase")
```

```python
recommendation_system('Spectre')
```
